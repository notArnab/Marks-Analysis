```python
X = cse_df.iloc[:,[6,7]].values
X.shape
```

```python
from  sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
  km = KMeans(n_clusters = i, init = 'k-means++' , max_iter=200 , n_init=10 , random_state=0)
  km.fit(X)
  wcss.append(km.inertia_)

plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('No Of Clusters')
plt.ylabel('wcss')
plt.show()
```

```python
km = KMeans(n_clusters = 2 , init = 'k-means++' , max_iter = 200 , n_init = 10 , random_state = 0)
y_means = km.fit_predict(X)
plt.scatter(X[y_means == 0,0] , X[y_means == 0,1] , s = 100 , c='pink' , label = 'Poor Student')
plt.scatter(X[y_means == 1,0] , X[y_means == 1,1] , s = 100 , c='yellow' , label = 'Good Student')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:, 1], s = 50, c = 'blue' , label = 'centeroid')

plt.xlabel('2nd_Sem', fontsize = 15)
plt.ylabel('3rd_Sem ', fontsize = 15)
plt.legend()
plt.show()
```

```python
cse1_df = cse_df.drop(columns=['University  Roll Number','FIRST NAME','STREAM'])
```

```python
y = cse1_df['4th_Sem_Marks']
X = cse1_df.drop('4th_Sem_Marks',axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X ,y , test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_pred,y_test)
print('RMSE Score :',np.sqrt(mse))

r2 = r2_score(y_pred,y_test)
print('R2 Score :',r2)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train,y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_pred,y_test)
print('RMSE Score :',np.sqrt(mse))

r2 = r2_score(y_pred,y_test)
print('R2 Score :',r2)
```

```python
from sklearn.svm import SVR
model = SVR()
model.fit(X_train,y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_pred,y_test)
print('RMSE Score :',np.sqrt(mse))

r2 = r2_score(y_pred,y_test)
print('R2 Score :',r2)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(X_train,y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_pred,y_test)
print('RMSE Score :',np.sqrt(mse))

r2 = r2_score(y_pred,y_test)
print('R2 Score :',r2)
```
